```python
def remove_col_train(col_name, X_train=X_train, X_val=X_val, y_train=y_train, y_val=y_val):

    X_train = X_train.drop(col_name, axis=1)
    X_val = X_val.drop(col_name, axis=1)

    dv = DictVectorizer(sparse=False)
    
    train_dict = X_train.to_dict(orient='records')
    val_dict = X_val.to_dict(orient='records')
    dv.fit(train_dict)
    X_train = dv.fit_transform(train_dict)
    X_val = dv.transform(val_dict)
    model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict_proba(X_val)[:, 1]
    conversion = (y_pred >= 0.5).astype(int)
    return (conversion == y_val).mean()
```

```python
acc = remove_col_train('industry')
acc - base_val_pred
```

```python
acc = remove_col_train('employment_status')
acc - base_val_pred
```

```python
acc = remove_col_train('lead_score')
acc - base_val_pred
```

## Question 6

* Now let's train a regularized logistic regression.
* Let's try the following values of the parameter `C`: `[0.01, 0.1, 1, 10, 100]`.
* Train models using all the features as in Q4.
* Calculate the accuracy on the validation dataset and round it to 3 decimal digits.

Which of these `C` leads to the best accuracy on the validation set?

```python
def train(C, X_train=X_train, X_val=X_val, y_train=y_train, y_val=y_val):

    dv = DictVectorizer(sparse=False)
    
    train_dict = X_train.to_dict(orient='records')
    val_dict = X_val.to_dict(orient='records')
    dv.fit(train_dict)
    X_train = dv.fit_transform(train_dict)
    X_val = dv.transform(val_dict)
    model = LogisticRegression(solver='liblinear', C=C, max_iter=1000, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict_proba(X_val)[:, 1]
    conversion = (y_pred >= 0.5).astype(int)
    return (conversion == y_val).mean()
```

```python
C_list = [0.01, 0.1, 1, 10, 100]

for C in C_list:
    print(f"{C}: {train(C)}")
```
